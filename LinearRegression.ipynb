# Predicting Movie Box office Revenue with Linear Regression

# Import Libraries

```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# Import the dataset

```python
data= pandas.read_csv('cost-revenue-clean.csv')
```

# Understanding the Data

```python
data.head()
```

```python
data.describe()
```

# Analysis of Data using Linear Regression

```python
#Splitting the dataFrame into feature and target variable
X=DataFrame(data,columns=['production_budget_usd'])
y=DataFrame(data,columns=['worldwide_gross_usd'])
```

# Visualization of Data Using Scatter Chart

```python
#Visualization required for the data
plt.figure(figsize=(10,6))
#To add transparency in the chart
plt.scatter(X,y,alpha=0.3)  #this gives us information about the data distribution
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('WorldWide Gross $')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

# Linear Regression using scikit-learn

```python
regression=LinearRegression()
regression.fit(X,y)
```

# Slope Coefficient

```python
regression.coef_ #theta
```

```python
#Intercept
regression.intercept_
```

```python
# Plotting the regression Line 
#Visualization required for the data
plt.figure(figsize=(10,6))
#To add transparency in the chart
plt.scatter(X,y,alpha=0.3)  #this gives us information about the data distribution
#Plot a regression line on chart
plt.plot(X,regression.predict(X),color='red',linewidth=4)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('WorldWide Gross $')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

# Inferences

#### 1) From the regression.coef_ we come to realize that there is positive correlation between production budget and WorldWide gross.Also physical significance is that for every 1$ spent on production budget we in turn make up to 3.11 in WorldWide Gross.So it seems that our model is explaining our intution that feature variable(Production Budget)and WorldWide Gross has some relationship.
2)Coming back to the second parameter i.e constant .It tells that a movie with budget of $0 we will loose 7236192.So we should take our model with a grain of salt.

```python
regression.score(X,y) #R squared
# This parameter suggests the amount of variation in film revenue that is explained by film's budget.
```
